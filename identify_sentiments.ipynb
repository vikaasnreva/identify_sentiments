```python
import nltk.corpus
from textblob import TextBlob
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv",)
```

### Data pre-processing

```python
train=train.dropna(axis=0,how='any')
test=test.dropna(axis=0,how='any')
```

```python
def removetext(text):
    return ''.join([i if ord(i) < 128 else '' for i in text])

#Here I am doing the actual removing
train['tweet'] = train['tweet'].apply(removetext)

#Make all my texts lower case
train['tweet'] = train['tweet'].apply(lambda x: x.lower())
#for test data
test['tweet'] = test['tweet'].apply(removetext)

#Make all my texts lower case
test['tweet'] = test['tweet'].apply(lambda x: x.lower())
```

```python
#Get rid of all weird punctuation and extra lines
train['tweet'] = train['tweet'].apply(lambda x: x.replace('.',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace('\n',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace('?',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace('!',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace('"',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace(';',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace('#',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace('&amp',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace(',',' '))
train['tweet'] = train['tweet'].apply(lambda x: x.replace('^^',' '))
 #for test data
test['tweet'] = test['tweet'].apply(lambda x: x.replace('.',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace('\n',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace('?',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace('!',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace('"',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace(';',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace('#',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace('&amp',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace(',',' '))
test['tweet'] = test['tweet'].apply(lambda x: x.replace('^^',' '))
```

```python
stopwords = set(STOPWORDS)
```

```python
s=[]
print('Tweet   |     Polarity     |     Subjectivity')
for val in test.tweet:
    sentiments=TextBlob(val)
    s.append(sentiments.polarity)
    print('---------------')
    print(val,end='')
    print(' ',end='')
    print(sentiments.polarity,end='')
    print(' |  ',end='')
    print(sentiments.subjectivity)
```
